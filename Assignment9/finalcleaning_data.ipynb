# Cleaning SEU Dataset

```python
import pandas as pd
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px

df_Degree = pd.read_csv('DegreeConferrals(OG).csv')
df_Person = pd.read_csv('Person(OG).csv')
df_PersonDemographics = pd.read_csv('PersonDemographics(OG).csv')

df_Degree.head()
```

### Cleaning Major1_Desc Text

```python
sorted_unique_major1_desc = sorted(df_Person['Major1_Desc'].unique())
for major_desc in sorted_unique_major1_desc:
    print(major_desc)

df_Person['Major1_Desc'] = df_Person['Major1_Desc'].apply(lambda x: x.lower().title())
```

### Dropped Unknown Gender Value

```python
df_PersonDemographics = df_PersonDemographics[df_PersonDemographics['Gender'] != 'Uknown' ]
```

### Cleaned Citizenship Country Text

```python
sorted_unique_citizenship_country = sorted(df_PersonDemographics['CitizenshipCountry'].unique())
def process_citizenship_country(value):
    return value.strip().lower().title()
df_PersonDemographics['CitizenshipCountry'] = df_PersonDemographics['CitizenshipCountry'].apply(process_citizenship_country)
```
